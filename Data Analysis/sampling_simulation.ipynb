```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/content/am_coeffs_10240.csv", index_col=False)
df
```

```python
denom = df.values[:, 2]
num  = df.values[:, 1]
assert(len(denom) == len(num))
for k in range(len(denom)):
  if int(num[k]) == 0:
    denom[k] = 0
first_digits_denom = list(map(lambda p:str(abs(int(p)))[0], denom[:]))
first_digits_num = list(map(lambda p:str(abs(int(p)))[0], num[:]))
print(len(first_digits_num))
print(len(first_digits_denom))
```

```python
#first digit of a number
import math 
def ndigit(num, n):
    if num == 0 :
        return 0
    elif num < 0:
        num *= -1
    while((num<1) & (num>0)):
        num*=10
    return num // 10 ** (int(math.log(num, 10)) - n + 1)

decimals = []
for k, e in enumerate(df.values[:, 1]):
  if int(e) != 0:
    decimals.append(int(e)/int(denom[k]))
  else:
    decimals.append(0)
#decimals = [int(e)/int(denom[k]) for k, e in enumerate(df.values[:, 1]) if int(e) != 0]
first_digits_dec = [int(ndigit(decimals[i], 1)) for i in range(len(decimals))]
print(len(first_digits_dec))
```

```python
from collections import Counter

denum_counter = Counter(first_digits_denom)
num_counter = Counter(first_digits_num)
dec_counter = Counter(first_digits_dec)

col1 = [num_counter[f"{i}"] for i in range(1, 10)]
col2 = [denum_counter[f"{i}"] for i in range(1, 10)]
col3 = [dec_counter[i] for i in range(1, 10)]
final = {
    "numerator":col1,
    "denominator":col2,
    "decimal":col3
}
print(sum(col1), sum(col2), sum(col3))
final_df = pd.DataFrame(final, index=range(1, 10))
final_df
```

```python
import random
from scipy.stats import chisquare
random.seed(11)

threshold = 15.5073
benford_distribution = [0.30103, 0.176091, 0.124939, 0.09691, 0.079181, 0.066947, 0.057992, 0.051153, 0.045757]

def find_chisquare(first_digits):
  benford_distribution = [0.30103, 0.176091, 0.124939, 0.09691, 0.079181, 0.066947, 0.057992, 0.051153, 0.045757]
  list_of_interest = list(map(int, first_digits))
  counter = Counter(list_of_interest)
  observed = [counter[i] for i in range(1, 10)]
  return chisquare(observed, f_exp=[sum(observed)*i for i in benford_distribution]) # chi square, p-value

count = 0

for _ in range(1000):
    sampled = list(map(int, random.sample(first_digits_num, 1000)))
    if find_chisquare(sampled)[0] > threshold:
      count += 1

print(count)
```
